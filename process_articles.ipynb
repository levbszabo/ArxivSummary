```
def get_data(in_path, out_path, n=10000):
    article_dicts = []
    file_in = open(in_path,"r")      
    lines = file_in.readlines()
    print("Number of articles are:", len(lines))
    lines_subsets = lines[:n]
    # for f in lines_subsets:
    #     article_dicts.append(json.loads(f))
    file_in.close()
    # with open(out_path, 'w') as filehandle:
    #   json.dump(lines_subsets, filehandle)
    with open(out_path, 'w') as filehandle:
      filehandle.writelines("%s" % line for line in lines_subsets)
    return article_dicts
  
```
